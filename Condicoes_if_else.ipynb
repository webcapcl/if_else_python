```python
entrega = input('Digite o tipo de entrega: ')
if entrega == 'padrão':
  print('Entrega padrão: R$ 10.00')
elif entrega == 'expresso':
  print('Entrega expresso: R$ 20.00')
else:
  print('Opção e entrega inválida')
```

```python
# Atividade 01

numero = float(input('Digite um número: '))
if numero > 0:
  print('Número positivo')
elif numero == 0:
  print('Númeo nulo')
else:
  print('Número negativo')
```

```python
# Atividade 02

numero = int(input('Digite seu númeo: '))
if numero % 2 == 0:
  print('O número é par')
else:
  print('O númeo e impar')
```

```python
# Atividade 03

nota = float(input('Digite sua nota: '))
if nota >= 7:
  print('Aprovado')
elif nota >= 5:
  print('Recuperação')
else:
  print('Reprovado')
```
